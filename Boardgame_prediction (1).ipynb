```
import pandas
import sklearn
import sys
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
#load data
games=pandas.read_csv("/content/games.csv")
#check data
#histogram average_rating of all the games
plt.hist(games["average_rating"])
plt.show()
```

```
#remove rows wothout user reviews
games=games[games["users_rated"]>0]
#removes rows with missing values
games=games.dropna(axis=0)
plt.hist(games["average_rating"])
plt.show()
```

```
#cheeck for correleation
corr=games.corr()
fig=plt.figure(figsize=(12,10))
sns.heatmap(corr,vmax=0.8,square=True)
```

```
coloumns=games.columns.to_list()
coloumns=[c for c in coloumns if c not in ["bayes_average_rating","average_rating","type","name","id"]]
#target of prediction
target="average_rating"
```

```
#split dataset for training and testing
#generate training set
train=games.sample(frac=0.8,random_state=1)

test=games.loc[~games.index.isin(train.index)]

print(train.shape)
print(test.shape)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr=LinearRegression()
#fit data into models
lr.fit(train[coloumns],train[target])
```

```
#generate predictions 
predictions=lr.predict(test[coloumns])
#value of error between predictions
mean_squared_error(predictions,test[target])
```

```
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=100,random_state=1,min_samples_leaf=10)
rfr.fit(train[coloumns],train[target])

```

```
predictions=rfr.predict(test[coloumns])
mean_squared_error(predictions,test[target]) 
```
